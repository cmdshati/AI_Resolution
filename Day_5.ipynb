# **Draw Multiple  Line**


```python
from matplotlib.lines import lineStyles
x = [1,2,3,4,5,6]
y1 = [7,8,9,5,2,12]
y2 = [4,9,36,78,9,10]
plt.plot(x,y1, label="Line 1", color = "green", marker = "s", linewidth= 2, linestyle= ":")
plt.plot(x,y2, label = "Line 2", color = "purple", marker = "o", linewidth = 2, linestyle="-.")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Multiple Line Graph")
plt.legend()
plt.show()
```

# **Draw Bar graph **

```python
player = ["Sakib", "Mushfiq", "Tamim", "Nasir", "Riyad"]
runs = [1000, 700, 500,600, 900]
plt.bar(player, runs, color = "green")
plt.xlabel("Player")
plt.ylabel("Runs")
plt.title("Bar Graph")
plt.show()
```

# Creating Horizontal Bar Plot

```python
player = ["Sakib", "Mushfiq", "Tamim", "Nasir", "Riyad"]
runs = [1000, 700, 500,600, 900]
plt.barh(player, runs, color = "pink")
plt.xlabel("Player")
plt.ylabel("Runs")
plt.title("Player vs Runs")
plt.show()
```

```python
player = ["Sakib", "Mushfiq", "Tamim", "Nasir", "Riyad"]
runs = [1000, 700, 500,600, 900]
plt.bar(player, runs, color = "pink", width = 0.5)
plt.xlabel("Player")
plt.ylabel("Runs")
plt.title("Player vs Runs")
plt.show()
```

# Draw Histogram

```python
runs = [ 2,4,5, 6, 2, 4, 6, 7, 8, 5, 9, 10, 3, 4 , 6, 10, 2, 3, 5, 6, 8, 4,0,0,1, 2, 2, 10]
plt.hist(runs, color = "green" , edgecolor= "black")
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
import numpy as np
data = np.random.randn(100)
plt.hist(data, color = "skyblue", edgecolor = "black", bins = 8)
plt.xlabel("Random Data")
plt.ylabel("Frequency")
plt.title("Histogram of Random Data")
plt.show()
```

# Draw Pie Chart (To show dataset is balanced or not?)

```python
Food = ["Protein", "Fat", "Water", "Vitamin"]
Percentage = [40, 20,50,10]
plt.pie(Percentage,labels = Food,autopct='%1.1f%%', shadow=True, startangle=90, explode=(0,0,0.1,0))
plt.title("Basic Pie Chart")
plt.show()
```

```python
Food = ["Protein", "Fat", "Water", "Vitamin"]
Percentage = [40, 20, 30, 10]

# explode = (0.2, 0, 0, 0)
explode = (0, 0.2, 0.1, 0)

plt.pie(Percentage,labels=Food,explode=explode,autopct='%1.1f%%',shadow=True,startangle=140)
plt.show()
```

# Multiple figure draw in a row

```python
x = [1,2,3,4,5,6]
y1 = [7,8,9,10,11,12]
y2 = [4,9,36,78,9,10]


#first plot
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line Plot")

#second plot
plt.subplot(1,2,2)
plt.plot(y1,y2)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line Plot")
```
