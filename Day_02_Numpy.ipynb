# Copy Array :Method 1 :Normal Assignment , Method 2 : Copy  , Method 3  : view

```python
list = [1,2,3,45]
arr1 = np.array(list)

# Normal Assignment
new_array = arr1
# before change
print("before change of new_array")
print(new_array)
print(arr1)

new_array[0]=100
print(new_array)
print(arr1)
if(id(arr1)==id(new_array)):
    print("same")
else:
    print("not same")



```

# Method  2 : Copy

```python
list = [1,2,3,4]
arr1 = np.array(list)

arr2 = np.copy(arr1)

print(arr1)
print(arr2)

arr2[3]=67

if(id(arr1)==id(arr2)):
    print("same")
else:
    print("not same")

print(arr1)
print(arr2)

```

# Method 3 View

```python
list = [1,2,3,4]
arr1 = np.array(list)

arr2 = arr1.view()

print(arr1)
print(arr2)

arr2[3]=67

if(id(arr1)==id(arr2)):
    print("same")
else:
    print("not same")

print(arr1)
print(arr2)

```

## Indexing, Slicing, Subsettng in an array

```python
a = np.array([[1,2,3],[8,9,0]])
print(a)
# indexing
print(a[1,0])

print(a[[0,0,1],[0,1,1]])

x =  np.array([1,9,0,7])
print(x[0])

x[0] = 8
print(x)

# Slicing
print(x[2:])
print(x[:-2])

print(x[-2:])
```

# Slicing in 2D Array

```python
a = np.array([[25,18,45,89,13],
              [90,78,56,34,67],
              [34,67,87,98,34]])
print(a)

# Get the second row (index 1)
print(a[1,:])
# get the third column ( index  2)
print(a[:, 2])

# get  all rows from 1 to last
print(a[1:,])

# Extracting Submatrices
#  Get rows 0 to 1, and columns 0 to 2
print(a[0:2,0:3])

# negative  slicing
print(a[-1:,])
print(a[:,-1])

print(a[:,1])
```

# np.fromiter : Creates an array from an iterable object . It bulid only  1 Dimnesioanl array . To create 2D ar 3D array first create 1D array then reshape it using reshape method. It has parameters like (iterable object, dtype, count for specifying item)

```python
square = (x * x for x in range(5))

array = np.fromiter(square,dtype=int)
print(array)
```

```python
square = (x * x for x in range(10))

array = np.fromiter(square,dtype=int)
print(array)
```

```python
square = (x * x for x in range(10))

# use of count parameter
array = np.fromiter(square,dtype=int ,count=2)
print(array)
```

## Numpy array Manipulation

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
print(np.concatenate((arr1,arr2)))
print(np.concatenate((arr1,arr2),axis =0))

# Using stack creates 2d array from 1d array and creates 3D array from 2D array .
print(np.stack((arr1,arr2),axis =0))
print(np.stack((arr1,arr2),axis =1))



```

# stacking two 1D arrays will produce a 2D array, and stacking multiple 2D arrays will produce a 3D array

```python
arr1 = np.array([[1,2,3],[7,8,9]])
arr2 = np.array([[4,5,6],[0,9,5]])
print(np.concatenate((arr1,arr2),axis =0))
print(np.concatenate((arr1,arr2),axis =1))
print(np.concatenate((arr1,arr2),axis =None))
print(np.stack((arr1,arr2),axis =0))
print(np.stack((arr1,arr2),axis =1))
print(np.stack((arr1,arr2),axis =2))




```
