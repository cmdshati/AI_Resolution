# Combining Two Dataset

```python
DataFrame1 = pd.DataFrame({"Fruit": ["apple", "mango", "lichi", "Orange"], "Quantity":[20, 39, 60, 67], "Price":[400, 500, 600, 800]})
DataFrame1
```

```python
DataFrame2 = pd.DataFrame({"Fruit": ["apple", "Papaya", "Guava", "Orange"], "Quantity":[20, 39, 60, 67], "Price":[400, 500, 600, 800]})
DataFrame2
```

```python
mergedf = pd.merge(DataFrame1, DataFrame2, on="Fruit")
mergedf
```

```python
mergedf = pd.merge(DataFrame1, DataFrame2, on="Fruit", how="right")
mergedf
```

```python
mergedf = pd.merge(DataFrame1, DataFrame2, on="Fruit", how="left")
mergedf
```

```python
mergedf = pd.merge(DataFrame1, DataFrame2, on="Fruit", how="outer")
mergedf
```

Concat two data set

```python
dfconcat= pd.concat([DataFrame1, DataFrame2])
dfconcat
```

```python
dfconcatRow= pd.concat([DataFrame1, DataFrame2],ignore_index=True,axis=0)
# axis 0  means row wise concat
dfconcatRow
```

```python
dfconcatColumn= pd.concat([DataFrame1, DataFrame2],ignore_index=True,axis=1)
# axis 1 means comulumn wise concat
dfconcatColumn
```

# Adding new Column in the concat Dataset

```python
newData = {"Discount":[2, 6, 8, 9, 6, 9,3]}
# conver into DataFrame
Discount = pd.DataFrame(newData)
Discount

```

```python
newDataSet= pd.concat([dfconcatRow,Discount],axis=1)
newDataSet
```
