# Nan means Not  a Number . Used to represent missing number , uuncomputable number .

```python
import pandas as  pd
import numpy as np


data = {"Name": ["Anna", "Mike", "Sarah","Zarah", "John"],
        "Age":[28,np.nan,67,56,np.nan],
        "Salary":[6000,70000, np.nan, 56000,45000]}
df = pd.DataFrame(data)
df

# how to find missing values
df.isnull().sum()
```

# Handle Missing Value : Method 1 ( Remove Missing Values)

```python
df_drop =df.dropna()
df_drop
```

# Method 2 Fill with a constant value

```python
df_fill =df.fillna("Unknown")
df_fill
```

```python
df_fill =df.fillna(0)
df_fill
```

# Method 3 - Fill with a Statistical value like mean or median

```python
df_stat = df.copy()
df_stat['Age'] = df_stat['Age'].fillna(df_stat['Age'].mean())
df_stat['Salary'] = df_stat['Salary'].fillna(df_stat['Salary'].median())
df_stat
```

```python
df_stat = df.copy()
df_stat['Age'] = df_stat['Age'].fillna(df_stat['Age'].mean())
df_stat['Salary'] = df_stat['Salary'].fillna(df_stat['Salary'].median())
df_stat
```

# Method 4 : Fill with a Forward value also known as Forward Fill

```python
df_ffill = df.copy()
df_ffill['Age'] = df_ffill['Age'].fillna(method='ffill')
df_ffill['Salary'] = df_ffill['Salary'].fillna(method='ffill')
df_ffill
```

```python
df_ffill= df.fillna(method='ffill')
df_ffill
```

# Method 5 Fill with a backward value  Backward Fill

```python
df_bfill = df.copy()
df_bfill['Age'] = df_bfill['Age'].fillna(method='bfill')
df_bfill['Salary'] = df_bfill['Salary'].fillna(method='bfill')
df_bfill
```

```python
df_bfill= df.fillna(method='bfill')
df_bfill
```

# Method 6 Interpolation

```python
df_interpol = df.copy()
df_interpol['Age'] = df_interpol['Age'].interpolate()
df_interpol['Salary'] = df_interpol['Salary'].interpolate()
df_interpol
```

```python
df_interpol= df.interpolate()
df_interpol
```

```python
df_drop = df.drop_duplicates()
df_drop
```

# Data Transformation : Pivot Table

```python
data = {"Name": ["Anna", "Mike", "Sarah","Zarah", "Alex"],
        "Department":["HR","IT","HR","IT","HR"],
        "Gender":["Female","Male","Female","Female","Male"],
        "Marks":[80, 75, 89, 67, 90]}

# convert  it inoto dataframe
df = pd.DataFrame(data)
df
```

# To show marks in departmentwise  accroding to gender . Use Pivot table for that.

```python
pivotTable = df.pivot_table(index='Department', columns='Gender', values='Marks', aggfunc='mean')
pivotTable
```

```python
pivotTable = df.pivot_table(index='Department', columns='Gender', values='Marks', aggfunc='sum')
pivotTable
```
